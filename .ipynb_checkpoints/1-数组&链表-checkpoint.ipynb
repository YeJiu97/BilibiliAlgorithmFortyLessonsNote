第一题：

```python
# 中文地址：https://leetcode.cn/problems/reverse-linked-list/
# 题解：

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prev, curr = None, head  # curr指向head，以及之后表示当前
        while curr is not None:  # 当curr不为None的时候
            next = curr.next  # next变量指向curr的下一个结点，也就是curr.next
            curr.next = prev  # 让curr.next指向prev，比如最开始的时候就是将结点的head指向None（存储在prev）
            prev = curr  # prev指向curr所在的结点
            curr = next  # curr指向next，也就是curr.next结点
        return prev  # 将结果返回
```

```python
# 另一种解法

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        curr, prev = head, None
        while curr:
            curr.next, prev, curr = prev, curr, curr.next
        return prev
```

第二题：

```python
# 中文地址：https://leetcode.cn/problems/linked-list-cycle/
# 题解： 哈希表法

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        # 考虑到linked list为None或者linked list只有一个结点的情况
        if head is None or head.next is None:
            return False
        
        # 创建一个hash表格
        hash_table = {}

        while head:
            if hash_table.get(head):  # 确认一下hash表中是否已经被存储了
                return True
            hash_table[head] = head.val  # 存储到哈希表中
            head = head.next  # 让head指向head的下一个结点
        
        # 如果循环结束都没有return True，那么就return False
        return False
```

运行结果：



```python
# 题解：快慢指针

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, x):
#         self.val = x
#         self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:

        # 只有一个结点或者链表为None
        if head is None or head.next is None:
            return False

        cur, prev = head, head  # 创建两个变量，并且都指向head

        while cur.next and cur.next.next:  # 只有cur的下一个不是None并且cur的下下一个结点不是None的时候继续循环
            cur = cur.next.next  # cur会指向cur的下一个结点的下一个结点
            prev = prev.next  # prev只会只想prev的下一个结点

            if cur == prev:  # 如果出现了两者相等的情况
                return True
        
        # 如果有循环，那么无法跳出上面的那个循环
        return False
```

## 复杂度分析

几种常见的复杂度：

1. O(1)：常数复杂度
2. O(log n)：对数复杂度
3. O(n)：线性时间复杂度
4. O(n^2)：平方
5. O(n^3)：立方
6. O(2^n)：指数
7. O(n!)：阶乘

O(1)代码：

```python
n = 10
print(n)
```

对于这一段代码而言，n的值的改变是不会导致代码的执行次数的改变：

```python
n = 100
print(n)
```

无论n是增加了2倍，还是增加了10倍，都不会对代码的执行量产生影响。

O(n)代码：

```python
n = 10
for i in range(10):
    print(i)
```

这段代码打印了10个数字（我们不需要关心数字的具体的值，只需要关心打印的数字的数量）。

现在我们将n的值增加2倍，那么代码的执行次数也会相应的增加2倍：

```python
n = 20
for i in range(20):
    print(i)
```

现在打印的数字的数量为20个，也就是需要被打印的数字的数量增加了2倍。

如果输入数据的值增加n倍，代码的执行次数增加n倍，那么这个算法的复杂度为O(n)。

O(n^2)代码：

```python
n = 5
for i in range(n):
    for j in range(n):
        print(j)
```

上面的代码是一个循环套着一个循环，并且每一个循环都使用了n作为标记，所以最终执行的次数为n * n，也就是n^2。

同理，只需要我们在嵌套一层使用n作为标记的循环，那么复杂度就是O(n^3)。

O(log(n))代码：

```python
n = 4
flag = 1
while flag<n:
    print(flag)
    flag = flag * 2
```

我们接下来将n的值增加4倍：

```python
n = 16
flag = 1
while flag < n:
    print(flag)
    flag = flag * 2
```

数据量增加n倍，那么需要被执行的次数会增加log(n)倍。
